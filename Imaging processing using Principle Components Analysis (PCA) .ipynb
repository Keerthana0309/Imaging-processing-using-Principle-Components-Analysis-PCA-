```python

from pylab import imread,subplot,imshow,title,gray,figure,show,NullLocator
from ipykernel import kernelapp as app
#from pylab import *
from PIL import Image, ImageOps
%matplotlib inline

#A = Image.open('./images/hummingbirdcapsized.jpg')
A = Image.open("C:\\Users\\USER\\Desktop\\photos\\IMG-4025.JPG") # open an image
print(A.layers)
```

```python
print('Original Color Image')
imshow(A)
```

```python
A = mean(A,2) # to get a 2-D array, averaging along the 3nd axis.
print('Original Gray Image')
imshow(A)
```

```python
#for value 10
import matplotlib.pyplot as plt
full_pc = size(A,axis=1) # numbers of all the principal components
r = len(A[:,0])/len(A[1])
print(r,len(A[1]),A.shape,A.size,full_pc)

i = 1
dist = []
plt.figure(figsize=(11,11*r))
for numpc in range(0,full_pc+10,50): # 0 50 100 ... full_pc
    coeff, score, latent = princomp(A,numpc)
    print(numpc,'coeff, score, latent \n',coeff.shape, score.shape, latent.shape)
    Ar = dot(coeff,score).T+mean(A,axis=0) # image reconstruction (project back)
    # difference in Frobenius norm
    dist.append(linalg.norm(A-Ar,'fro'))
    # showing the pics reconstructed with less than 50 PCs
    if numpc <= 250:
        ax = subplot(2,3,i,frame_on=False)
        ax.xaxis.set_major_locator(NullLocator()) # remove ticks
        ax.yaxis.set_major_locator(NullLocator())
        i += 1 
        imshow(Ar)  #imshow(flipud(Ar))
        title('PCs # '+str(numpc))
        gray()

figure()
imshow(A) #imshow(flipud(A))
title('numpc FULL: '+str(len(A[1])))
gray()
show()
```

```python
from pylab import plot,axis
figure()
perc = cumsum(latent)/sum(latent)
dist = dist/max(dist)
plot(range(len(perc)),perc,'b',range(0,full_pc+10,50),dist,'r')
axis([0,full_pc,0,1.1])
show()
```
